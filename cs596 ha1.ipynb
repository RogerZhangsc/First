```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

#### 1.1 Plot of straight line

```python
x1 = np.arange(0, 5, 0.01)
y1 = 30 + 0.5*x1
plt.plot(x1, y1, lw =2.0)
plt.title('straight line')
```

#### 1.2 Plot of quadratic function

```python
x2 = np.arange(0, 50, 0.1)
y2 = (x2 - 25)**2 + 20
plt.plot(x2, y2, lw=2)
plt.title('quadratic function')
plt.ylim(0, 100)
```

#### 1.3 Plot of log

```python
x3 = np.arange(0.001, 10, 0.001)
y3 = -np.log(x3)

x4 = np.arange(-10, 1, 0.001)
y4 = -np.log(1-x4)
plt.plot(x3, y3, x4, y4)
plt.title('log function')
plt.legend(['y=-log(x)', 'y=-log(1-x)'])

```

#### 1.4 Plot the sigmoid function

```python
x5 = np.arange(-20, 20, 0.01)
y5 = 1/(1 + np.e**(-x5))
plt.plot(x5, y5)
plt.title('sigmoid function')
```
